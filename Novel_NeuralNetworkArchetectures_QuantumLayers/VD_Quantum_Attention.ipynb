```python
import cudaq
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Function
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import shap
from lime.lime_tabular import LimeTabularExplainer
```

```python
device = torch.device('cpu')
cudaq.set_target("qpp-cpu")
```

```python
# Define quantum operations

def ry(theta, qubit):
    cudaq.ry(theta, qubit)

def rx(theta, qubit):
    cudaq.rx(theta, qubit)
```

```python
class QuantumFunction(Function):
    def __init__(self, qubit_count: int, hamiltonian: cudaq.SpinOperator):
        @cudaq.kernel
        def kernel(qubit_count: int, thetas: np.ndarray):
            qubits = cudaq.qvector(qubit_count)
            # Simple example: apply RY, then RX to qubit 0
            ry(thetas[0], qubits[0])
            rx(thetas[1], qubits[0])

        self.kernel = kernel
        self.qubit_count = qubit_count
        self.hamiltonian = hamiltonian

    def run(self, theta_vals: torch.Tensor) -> torch.Tensor:
        """
        Execute the quantum circuit for all samples in theta_vals.
        theta_vals is shape [batch_size, 2] if qubit_count=2.
        """
        # Convert to numpy for CUDAQ
        theta_vals_np = theta_vals.cpu().numpy()
        # We need an integer qubit_count for each item in the batch
        qubit_count_list = [self.qubit_count] * theta_vals_np.shape[0]
        results = cudaq.observe(
            self.kernel,
            self.hamiltonian,
            qubit_count_list,
            theta_vals_np
        )
        # Gather the expectation values
        exp_vals = [res.expectation() for res in results]
        return torch.tensor(exp_vals, dtype=torch.float32).to(device)

    @staticmethod
    def forward(ctx, thetas: torch.Tensor, quantum_circuit, shift) -> torch.Tensor:
        """
        Forward pass: run the quantum circuit, store everything for backward.
        """
        ctx.shift = shift
        ctx.quantum_circuit = quantum_circuit
        exp_vals = ctx.quantum_circuit.run(thetas).reshape(-1, 1)
        ctx.save_for_backward(thetas, exp_vals)
        return exp_vals

    @staticmethod
    def backward(ctx, grad_output):
        """
        Backward pass: parameter-shift rule for each parameter.
        """
        thetas, _ = ctx.saved_tensors
        shift = ctx.shift
        quantum_circuit = ctx.quantum_circuit

        gradients = torch.zeros_like(thetas, device=device)
        for param_index in range(thetas.shape[1]):
            # + shift
            thetas_plus = thetas.clone()
            thetas_plus[:, param_index] += shift
            exp_vals_plus = quantum_circuit.run(thetas_plus)

            # - shift
            thetas_minus = thetas.clone()
            thetas_minus[:, param_index] -= shift
            exp_vals_minus = quantum_circuit.run(thetas_minus)

            # Finite difference derivative
            gradients[:, param_index] = (exp_vals_plus - exp_vals_minus) / (2.0 * shift)

        return gradients * grad_output, None, None

# -----------------------------------------------------------------------------
# Define the Quantum Layer
# -----------------------------------------------------------------------------
class QuantumLayer(nn.Module):

    def __init__(self, qubit_count: int, hamiltonian, shift: torch.Tensor):
        super(QuantumLayer, self).__init__()
        self.quantum_circuit = QuantumFunction(qubit_count, hamiltonian)
        self.shift = shift

    def forward(self, input: torch.Tensor):
        # We apply the staticmethod from the QuantumFunction class
        return QuantumFunction.apply(input, self.quantum_circuit, self.shift)


class SimpleSelfAttention(nn.Module):
    def __init__(self, embed_dim):
        super(SimpleSelfAttention, self).__init__()
        self.embed_dim = embed_dim
        self.W_q = nn.Linear(embed_dim, embed_dim)
        self.W_k = nn.Linear(embed_dim, embed_dim)
        self.W_v = nn.Linear(embed_dim, embed_dim)
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, x):
        Q = self.W_q(x)
        K = self.W_k(x)
        V = self.W_v(x)
        # Scaled dot-product attention
        scores = torch.bmm(Q, K.transpose(1, 2)) / (self.embed_dim ** 0.5)
        attn_weights = self.softmax(scores)
        out = torch.bmm(attn_weights, V)
        return out, attn_weights
```

```python
#   Hybrid Model with a classical front-end + optional attention + quantum layer

class Hybrid_QNN(nn.Module):
    def __init__(self, input_dim=8):
        super(Hybrid_QNN, self).__init__()

        # Classical front-end
        self.fc1 = nn.Linear(input_dim, 256)
        self.fc2 = nn.Linear(256, 128)
        self.dropout = nn.Dropout(0.25)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, 32)

        self.attn = nn.MultiheadAttention(embed_dim=32, num_heads=1, batch_first=True)

        # Additional linear after attention to reduce dimension if needed
        self.fc5 = nn.Linear(32, 2)

        # Quantum layer with 2 qubits
        self.quantum = QuantumLayer(
            qubit_count=2,
            hamiltonian=cudaq.spin.z(0),
            shift=torch.tensor(np.pi / 2)
        )

    def forward(self, x: torch.Tensor):
        # Pass through classical layers
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = torch.relu(self.fc3(x))
        x = torch.relu(self.fc4(x))
        x = x.unsqueeze(1)

        x_attn, attn_weights = self.attn(x, x, x)
        x = x_attn.squeeze(1)

        # Final linear before quantum
        x = torch.relu(self.fc5(x))  # -> [batch_size, 2]
        return torch.sigmoid(self.quantum(x)).view(-1)
```

```python
# Load and Preprocess the Data

import pandas as pd
import numpy as np

def load_data(path):
    data = pd.read_csv(path)
    return data

dataset_path = "/content/tablea1.csv"
data = load_data(dataset_path)

# Identify only the numeric columns
numeric_cols = data.select_dtypes(include=[np.number]).columns

# Fill missing values in numeric columns with their column-wise mean
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())

# Choose features and target
features = ['logM1/2', 'logRe', 'logAge', '[Z/H]', 'logM*/L', 'DlogAge', 'D[Z/H]', 'DlogM*/L']
target = 'logsigmae'

# Prepare X and y
X = data[features].values
y = data[target].values

# Normalize the features (and target if needed)
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

# Convert to Torch tensors
X = torch.tensor(X, dtype=torch.float32).to(device)
y = torch.tensor(y, dtype=torch.float32).to(device)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.1,
    random_state=42
)
```

```python
# Instantiate & Train Model

model = Hybrid_QNN(input_dim=len(features)).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.01)
loss_function = nn.MSELoss().to(device)

epochs = 10
losses = []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    y_hat_train = model(X_train)
    loss = loss_function(y_hat_train, y_train)
    loss.backward()
    optimizer.step()

    losses.append(loss.item())
    print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss.item():.6f}")
```

```python
# 5) Machine Unlearning Step

threshold = 0.05

# Evaluate model on the entire training set
model.eval()
with torch.no_grad():
    preds_train = model(X_train)

# Compute absolute error
abs_error = torch.abs(preds_train - y_train)

# Boolean mask for samples we keep (error <= threshold)
mask = abs_error <= threshold

# Filter out high-error samples
removed_count = (~mask).sum().item()
X_train_unlearned = X_train[mask]
y_train_unlearned = y_train[mask]

print(f"\n--- Machine Unlearning ---")
print(f"Threshold: {threshold}")


# 6) Re-train the model on the “unlearned” dataset (phase 2)

epochs_unlearning = 10
losses_unlearning = []

print("\n=== Training (Phase 2) on Unlearned Dataset ===")
for epoch in range(epochs_unlearning):
    model.train()
    optimizer.zero_grad()

    y_hat_unlearned = model(X_train_unlearned)
    loss_unlearn = loss_function(y_hat_unlearned, y_train_unlearned)
    loss_unlearn.backward()
    optimizer.step()

    losses_unlearning.append(loss_unlearn.item())
    print(f"Epoch {epoch + 1}/{epochs_unlearning}, Loss: {loss_unlearn.item():.6f}")
```

```python
# Evaluation

model.eval()
with torch.no_grad():
    y_hat_test = model(X_test)
    test_loss = loss_function(y_hat_test, y_test).item()

r2_score_value = -r2_score(y_test.cpu(), y_hat_test.cpu())
mae_score_value = mean_absolute_error(y_test.cpu(), y_hat_test.cpu())
rmse_score_value = np.sqrt(mean_squared_error(y_test.cpu(), y_hat_test.cpu()))
mse_score_value = mean_squared_error(y_test.cpu(), y_hat_test.cpu())

print("\n--- Evaluation Metrics ---")
print(f"R^2 Score: {r2_score_value:.4f}")
print(f"MAE:       {mae_score_value:.4f}")
print(f"RMSE:      {rmse_score_value:.4f}")
print(f"MSE:       {mse_score_value:.4f}")
```

```python
# Range in the scaled domain
y_min = torch.min(y_test).item()
y_max = torch.max(y_test).item()
data_range = y_max - y_min

# Accuracy = (1 - error / range) * 100
mae_accuracy = (1.0 - mae_score_value / data_range) * 100
rmse_accuracy = (1.0 - rmse_score_value / data_range) * 100
mse_accuracy = (1.0 - mse_score_value / data_range) * 100

print("\n--- Accuracy (based on scaled range) ---")
print(f"MAE Accuracy:  {mae_accuracy:.2f}%")
print(f"RMSE Accuracy: {rmse_accuracy:.2f}%")
print(f"MSE Accuracy:  {mse_accuracy:.2f}%")
```

```python
def model_predict(x_array):
    with torch.no_grad():
        x_tensor = torch.tensor(x_array, dtype=torch.float32, device=device)
        preds = model(x_tensor).cpu().numpy()  # shape [N]
    return preds
```

```python
import matplotlib.pyplot as plt

# 1) LIME Explanation
lime_explainer = LimeTabularExplainer(
    training_data=X_train.cpu().numpy(),
    feature_names=features,
    verbose=True,
    mode="regression",
    discretize_continuous=True,
)

# Explain a single test instance (index 0 for demonstration)
i = 0
test_instance = X_test[i].cpu().numpy()
lime_exp = lime_explainer.explain_instance(
    test_instance,
    model_predict,
    num_features=8
)

print("\n--- LIME Explanation for one test instance ---")
print(lime_exp.as_list())
```
